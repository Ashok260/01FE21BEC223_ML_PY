```python
import random
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
i = 0
c=0
u=0
fields=['round','comp_win','user_win']
round=[]
comp_won=[]
user_won=[]
while i!= 11:
    i+=1
    computer = random.randint(1, 9)
    userinput = int(input("Enter the number between 1 and 9"))
    print(computer)
    print(userinput)
    if computer > userinput:
        c = c + 1
    elif computer == userinput:
        print("tie")
    else:
        u=u+1
    round.append(i)
    comp_won.append(c/i)
    user_won.append(u/i)

rows=[[round[0],comp_won[0],user_won[0]],
      [round[1],comp_won[1],user_won[1]],
      [round[2],comp_won[2],user_won[2]],
      [round[3],comp_won[3],user_won[3]],
      [round[4],comp_won[4],user_won[4]],
      [round[5],comp_won[5],user_won[5]],
      [round[6],comp_won[6],user_won[6]],
      [round[7],comp_won[7],user_won[7]],
      [round[8],comp_won[8],user_won[8]],
      [round[9],comp_won[9],user_won[9]],
      [round[10],comp_won[10],user_won[10]]]
filename="record.csv"
with open(filename,'w') as csvfile:
    csvwriter=csv.writer(csvfile)
    csvwriter.writerow(fields)
    csvwriter.writerows(rows)
print(round)
print(comp_won)
print(user_won)
if c > u:
    print("computer won")
else:
    print("user won")
data=pd.DataFrame
data=pd.read_csv(r"/content/record.csv")
C=np.array(data.loc[0:10,("comp_win")])
U=np.array(data.loc[0:10,("user_win")])
R=np.array(data.loc[0:10,("round")])
print("Graph of success rate of computer")
plt.plot(C,color='Red',label='comp_success_rate')
plt.plot(U,color='Blue',label='user_success_rate')
plt.xlabel('Number of rounds')
plt.ylabel('success rate')
plt.title('user vs computer success rate')
plt.legend()
plt.grid()
plt.show()
```
